```python
aaa = {"a": {"2020-01-01": 2}, "b": {"2020-01-01": 2, "2020-01-02": 3}, "c": {"2020-01-01": 2, "2020-01-05": 3}}
aa = pd.DataFrame(aaa)
print(aaa, "\n", aa)
```

### æŸ¥çœ‹index å’Œ columns

```python
bb = aa.fillna(0)
print(bb.__dict__)
print(bb)
_index = [i for i in bb.index]
_cols = [i for i in bb.columns]
print(_index)
```

### éå†indexæŸ¥æ‰¾ç©ºç¼ºæ—¥æœŸ.
å¯æ ¹æ®`index list`æ’åºå, ç›¸é‚»ä¸¤æ—¥å·®å¤§äºä¸€å¤©çš„éƒ½åº”è¯¥è¡¥

<font color=green>éœ€è¦æ³¨æ„æ¯ä¸ª `a = a or b` ä¸­ å¾ªç¯éå†`a`çš„é‡ç½®åˆå§‹å€¼ä½ç½®.</font>

```python
import datetime
_last_date = ''
for _date in _index:
    _date = datetime.datetime.strptime(_date, "%Y-%m-%d")
    _last_date = _last_date or _date
    time_delta = (_date - _last_date).days
    list_need_add_day = []
    need_add_day = ''
    if time_delta > 1:
        for i in range(time_delta-1):
            need_add_day = need_add_day or _last_date
            need_add_day = need_add_day + datetime.timedelta(1)
            list_need_add_day.append(need_add_day)
    _last_date = _date
print(list_need_add_day)
```

### å‘ä¸»DataFrameä¸­å¹¶å…¥ç¼ºå°‘æ—¥æœŸ 
#### _coléšä¾¿æ’å…¥ä¸€ä¸ª. åç»­NaNè¡¥é›¶

```python
for i in list_need_add_day:
    i = i.strftime("%Y-%m-%d")
    aa = aa.append(pd.DataFrame({_cols[-1]: {i: 0}}))
    # print(aa)
bb = aa.fillna(0)
bb = bb.sort_index()
print(bb)
```
